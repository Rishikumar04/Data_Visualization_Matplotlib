## Matplotlib

Matplotlib is the "grandfather" library of data visualization with Python. It was created by John Hunter.

Matplotlib allows you to create reproducible figures programmatically. Explore the webpage:http://matplotlib.org/

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

### Basic Matplotlib plot



```python
plt.plot(x,y,'r')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Matplolib plot');
```

### Creating Subplot

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x, y, 'r--') # More on color options later
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-');
```

### Object Oriented Method

```python
#Creating figure
fig = plt.figure()

axes=fig.add_axes([0.1,0.1,0.5,0.5]) # left, bottom, width, height (range 0 to 1)

#Plot on that axes
axes.plot(x,y,'r')
axes.set_xlabel('Set X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Set y Label')
axes.set_title('Set Title');
```

```python
#Creating figure
fig = plt.figure()

axes1=fig.add_axes([0.1,0.1,0.9,0.9])
axes2=fig.add_axes([0.2,0.5,0.4,0.4])

axes1.plot(x,y,'g')
axes1.set_xlabel('X-axes-1')
axes1.set_ylabel('Y-axes-1')
axes1.set_title('Outer Plot')

axes2.plot(y,x,'r')
axes2.set_xlabel('X-axes-2')
axes2.set_ylabel('Y-axes-2')
axes2.set_title('Inner Plot')
```

#### Subplots
plt.subplots() will act as a automatic axis manager

```python
fig,axes = plt.subplots(nrows=1,ncols=3)
#to fix overlapping
plt.tight_layout()
```

```python
#indicates rows and columns of above plot
axes
```

```python
for ax in axes:
    ax.plot(x,y)
    ax.set_xlabel('x-axis')
    ax.set_ylabel('y-axis')
    ax.set_title('title')
#Display figure
fig
```

```python
#Another method
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x,'r');

```

#### Legend

```python
fig = plt.figure(figsize=(16,6))

ax = fig.add_axes([0.1,0.1,0.5,0.7])
ax.plot(x**2,label='x-square')
ax.plot(x**3,label='x-cube')
ax.legend(loc=0)
```

### Setting Colors

```python
fig=plt.figure(figsize=(16,6))

ax=fig.add_axes([0.1,0.1,0.6,0.6])
#can give lots of colors
ax.plot(x,y,c='green')
```

### Setting Linewidths, transparency, linestyle

```python
fig,axes=plt.subplots(nrows=1,ncols=3)
axes[0].plot(x,x**2,c='g',linewidth=3)#linewidths
axes[1].plot(x,x**2,c='r',linewidth=3,alpha=0.5)#alpha value will give transparency ranges from 0-1
axes[2].plot(x,x**2,c='b',linewidth=4,alpha=0.6,linestyle='-.')#various types of linestyle can be given.
plt.tight_layout()
```

### Marker

```python
#to markout each point
fig=plt.figure(figsize=(16,6))

axes1=fig.add_axes([0.1,0.1,0.6,0.7])

axes1.plot(x,y,color='c',linewidth=3,alpha=0.5,linestyle='--',marker='v',markerfacecolor='g',markeredgewidth=3,
           markeredgecolor='r',markersize=7)
```

### Plot Range

We can configure the ranges of the axes using the `set_ylim` and `set_xlim` methods in the axis object, or `axis('tight')` for automatically getting "tightly fitted" axes ranges:

```python
fig, axes= plt.subplots(nrows=1,ncols=3,figsize=(16,6))

axes[0].plot(x,x**2,x,x**3)
axes[0].set_title('Default axes range')

axes[1].plot(x,x**2,x,x**3)
axes[1].axis('tight')
axes[1].set_title('tight_axes')

axes[2].plot(x,x**2,x,x**3)
axes[2].set_ylim([0,60])
axes[2].set_xlim([2,5])
axes[2].set_title('Custom axes range')
```
